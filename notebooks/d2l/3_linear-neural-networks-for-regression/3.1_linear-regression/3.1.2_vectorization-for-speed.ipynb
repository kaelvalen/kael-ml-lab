# 3.1.2 Vectorization For Speed

```python
import torch
import time
```

```python
n = 10000
a = torch.ones(n)
b = torch.ones(n)
```

### benchmark

```python
c = torch.zeros(n)
t = time.time()
for i in range(n):
    c[i] = a[i] + b[i]
print(f"Time taken: {time.time() - t} seconds")
```

```python
t = time.time()
d = a + b
print(time.time() - t)
```
