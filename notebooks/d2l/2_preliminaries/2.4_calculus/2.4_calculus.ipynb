# Calculus and Its Role in Deep Learning

## Historical Context
For a long time, how to calculate the area of a circle remained a mystery. Then, in Ancient Greece, the mathematician Archimedes came up with the clever idea to inscribe a series of polygons with increasing numbers of vertices on the inside of a circle.

## Archimedes' Method
For a polygon with n vertices, we obtain n triangles. The height of each triangle approaches the radius r as we partition the circle more finely. At the same time, its base approaches 2r/n, since the ratio between arc and secant approaches 1 for a large number of vertices. Thus, the area of the polygon approaches n  r  (1/2)  (2r/n) = r.



## Connection to Calculus
This limiting procedure is at the root of both differential calculus and integral calculus.

## Applications in Deep Learning
- **Differential calculus** can tell us how to increase or decrease a function's value by manipulating its arguments
- This comes in handy for the **optimization problems** that we face in deep learning
- We repeatedly update our parameters in order to decrease the loss function
- **Optimization** addresses how to fit our models to training data, and calculus is its key prerequisite
