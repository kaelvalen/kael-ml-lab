# 2.5.4 Gradients And Python Control Flow

```python
import torch
```

```python
def f(a):
    b = a * 2
    while b.norm() < 1000:
        b = b * 2
    if b.sum() > 0:
        c = b
    else:
        c = 100 * b
    return c
```

```python
a = torch.randn(size=(), requires_grad=True)
d = f(a)
d
```

```python
d.backward()
```

```python
a.grad == d/a
```
