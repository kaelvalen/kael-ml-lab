# 2.5.2 Backward For Non Scalar Variables

```python
import torch
```

```python

x = torch.tensor([8.0], requires_grad=True)
y = x * 2

y.backward()
print(x.grad)
```

```python
x.grad.zero_()
```

```python
y = x * x
y.backward(gradient = torch.ones(len(y)))
```
