# 2.5.3 Detaching Computation

```python
# Let's set up our variables
import torch

x = torch.arange(4.0, requires_grad=True)
x
```

```python
# Create y = x * x
y = x * x
y
```

```python
# Create u by detaching y from the computation graph
u = y.detach()
u
```

```python
# Create z = x * u and compute gradients
z = u * x
z.sum().backward()
x.grad == u
```

Note that while this procedure detaches y's ancestors from the graph leading to z, the computational graph leading to y persists and thus we can calculate the gradient of y with respect to x.

```python
# We can still compute gradients for y with respect to x
x.grad.zero_()  # Clear previous gradients
y.sum().backward()
x.grad == 2 * x
```

```python
x = torch.arange(4.0)
x.requires_grad_(True)
print(x.grad)
```
