## 1.3.2 Unsupervised and Self-Supervised Learning

UnSupervised Learning:
Unsupervised learning aims to discover structure, patterns, and relationships in data without labels. Common tasks include clustering, dimensionality reduction (e.g., PCA), representation learning, causal discovery, and generative modeling. Deep generative models such as VAEs, GANs, normalizing flows, and diffusion models learn the data distribution and can generate new samples.

Self-Supervised Learning:
Self-supervised learning creates supervision signals from unlabeled data itself, such as predicting masked words in text or missing regions in images. These methods often learn strong representations that are later fine-tuned for downstream supervised tasks.

### Reminder
1. Supervised: labels tell you what to learn.
2. Unsupervised: data shows you what to learn.
3. Self-supervised: data teaches itself.
