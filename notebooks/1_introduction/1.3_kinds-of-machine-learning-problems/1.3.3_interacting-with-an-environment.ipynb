## 1.3.3 Interacting with an Environment

Interacting with an Environment:
Supervised and unsupervised learning are typically offline, meaning models are trained on fixed datasets without further interaction with the environment. However, intelligent systems must take actions, not just make predictions. Actions influence the environment and future observations, making offline learning insufficient.

When agents interact with environments, new challenges arise, such as memory, adversarial behavior, non-stationary dynamics, and distribution shift between training and deployment data. These problems are addressed within the framework of reinforcement learning, where an agent learns through interaction with an environment.
