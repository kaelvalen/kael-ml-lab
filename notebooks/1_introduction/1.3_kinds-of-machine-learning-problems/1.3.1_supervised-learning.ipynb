## 1.3.1 Supervised Learning

Supervised learning is a machine learning approach where models are trained on labeled data. Each example consists of input features and a corresponding label, and the goal is to learn a function that predicts the label from the features.

The process involves collecting labeled training data, using it to train a model, and then applying the trained model to unseen inputs to make predictions. Many real-world applications—such as medical diagnosis, machine translation, and price prediction—are framed as supervised learning problems because they can be expressed as predicting an unknown outcome given known data.

### Regression

Regression is one of the simplest supervised learning tasks. Each example is represented by a fixed-length feature vector (e.g., house size, number of rooms, distance to city center), and the goal is to predict a continuous numerical value.

A problem is called regression when the target (label) can take arbitrary numeric values, such as house prices, movie ratings, hospital stay duration, or rainfall amount. In general, questions like “how much?” or “how many?” indicate regression.

The objective is to learn a model whose predictions are as close as possible to the true values. When exact prediction is not possible due to noise or missing factors, models are trained to minimize prediction error, commonly using squared error loss. Linear regression is a basic example, assuming a linear relationship between features and the target and often Gaussian noise in the data.

### Classification

Classification is a supervised learning task where the goal is to predict which class an input belongs to from a discrete set of categories. Instead of predicting a numerical value (as in regression), a classifier outputs a class label, often expressed as probabilities over classes. Problems can be binary (two classes) or multiclass (more than two). Classification models are commonly trained using cross-entropy loss, and decisions may depend not only on the most likely class but also on the risk and consequences of errors.

### Tagging

Tagging is a classification problem where an example can belong to multiple classes at the same time. Unlike binary or multiclass classification, labels are not mutually exclusive. Common applications include image tagging, blog post tagging, and document indexing. Labels often show correlations, and real-world systems may involve very large label sets. Machine learning is often used to generate provisional tags to support search and organization.

### Search (Ranking)
In search problems, the goal is not just to decide whether an item is relevant, but to rank relevant items by importance. Models assign a score to each candidate and return the top-ranked results. Early systems like PageRank provided query-independent scores based on authority, while modern search engines use machine learning to produce query-dependent relevance rankings using user behavior and contextual signals.

### Recommender Systems
Recommender systems aim to present personalized items to users and are closely related to search and ranking. Unlike search, the focus is on user-specific relevance. Models are trained using explicit feedback (ratings, reviews) or implicit feedback (clicks, skips, purchases) to estimate scores such as expected ratings or purchase probability. Major challenges include censored feedback, bias, and feedback loops, where recommendations influence future user behavior and data.

Search = relevance, Recommendation = relevance + personalization.

### Sequence Learning:
Sequence learning deals with problems where inputs and/or outputs are ordered sequences and dependencies across time matter. Unlike fixed-size problems, observations are not independent, and models must retain context. Common examples include video analysis, language processing, and medical time-series monitoring.

Sequence-to-sequence learning handles variable-length input and output sequences, such as machine translation, speech recognition, and text-to-speech. Key challenges include unaligned sequences, different sequence lengths, and long-range dependencies.

Fixed ML forgets context — Sequence learning remembers the past.
