```python
#truth table of and
print("AND truth table")
print("A\tB\tA and B")
for A in [0, 1]:
    for B in [0, 1]:
        print(f"{A}\t{B}\t{A and B}")
```

```python
# truth table of or
print("a\tb\ta or b")
for a in [0, 1]:
    for b in [0, 1]:
        print(f"{a}\t{b}\t{a or b}")
```

```python
# not truth table
for i in range(2):
        print(i, not i)
```

```python
def crazy(n):
    count = 0

    total = n ** (n ** n)

    for i in range(total):
        count += 1

    return count
```
