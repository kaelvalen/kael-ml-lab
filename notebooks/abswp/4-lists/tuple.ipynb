```python
grid = [['.', '.', '.', '.', '.', '.'],        ['.', 'O', 'O', '.', '.', '.'],        ['O', 'O', 'O', 'O', '.', '.'],        ['O', 'O', 'O', 'O', 'O', '.'],        ['.', 'O', 'O', 'O', 'O', 'O'],        ['O', 'O', 'O', 'O', 'O', '.'],        ['O', 'O', 'O', 'O', '.', '.'],        ['.', 'O', 'O', '.', '.', '.'],        ['.', '.', '.', '.', '.', '.']]
```

```python
grid
```

```python
for y in range(len(grid[0])):  # Iterate over columns
    for x in range(len(grid)):  # Iterate over rows
        print(grid[x][y], end=' ')  # Print each character without a newline
    print()  # Print a newline after each column is finished
```
