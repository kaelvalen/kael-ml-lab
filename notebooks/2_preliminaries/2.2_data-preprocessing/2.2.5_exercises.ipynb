1. Try loading datasets, e.g., Abalone from theUCI Machine Learning Repository50andinspect their properties. What fraction of them has missing values? What fraction ofthe variables is numerical, categorical, or text?

```python
import pandas as pd
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data"
column_names = [
    "Sex", "Length", "Diameter", "Height", "WholeWeight",
    "ShuckedWeight", "VisceraWeight", "ShellWeight", "Rings"
]
```

```python
abalone_data = pd.read_csv(url, header=None, names=column_names)
```

```python
missing_values_fraction = abalone_data.isnull().mean()
```

```python
data_types = abalone_data.dtypes
```

```python
num_numerical = sum(data_types == "float64") + sum(data_types == "int64")
num_categorical = sum(data_types == "object")
num_text = 0
```

```python
total_columns = len(column_names)
fraction_numerical = num_numerical / total_columns
fraction_categorical = num_categorical / total_columns
fraction_text = num_text / total_columns
```

```python


# Print results
print("Fraction of missing values in each column:")
print(missing_values_fraction)
print("\nFraction of variables:")
print(f"Numerical: {fraction_numerical:.2f}")
print(f"Categorical: {fraction_categorical:.2f}")
print(f"Text: {fraction_text:.2f}")
```
