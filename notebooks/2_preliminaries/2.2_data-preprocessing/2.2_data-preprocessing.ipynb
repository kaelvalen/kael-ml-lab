## 2.2 Data Preprocessing
### Working with Real-World Data and Preprocessing

Until now, we have used synthetic data in ready-made tensors. However, real-world deep learning applications require handling messy data stored in various formats and preprocessing it to fit our needs.  

The `pandas` library is a powerful tool that simplifies much of this preprocessing. While this section is not a full `pandas` tutorial, it provides a quick overview of some of the most common routines for working with data.
