# 2.1.6 Conversion To Other Python Objects

```python
import numpy as np
import torch
```

Converting to a NumPy tensor (ndarray), or vice versa, is easy. The torch tensor andNumPy array will share their underlying memory, and changing one through an in-placeoperation will also change the other.

```python
x = torch.tensor([[0.0, 1.0], [2.0, 3.0]])
```

```python
A = x.numpy()
B = torch.from_numpy(A)
type(B), type(A)
```
