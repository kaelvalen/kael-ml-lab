### Scalars in Mathematics and Tensors

In everyday mathematics, we often work with single values called **scalars**.  
- Example: The temperature in Palo Alto is 72Â°F. To convert it to Celsius, we use the formula:  
  $c = \frac{5}{9}(f - 32)$, where $f = 72$.  
  - Here, 5, 9, and 32 are constant scalars, while $c$ and $f$ are variables representing unknown scalars.  

#### Notation
- Scalars are denoted by lowercase letters (e.g., $x$, $y$, $z$).  
- The space of all real-valued scalars is denoted by $\mathbb{R}$.  
  - For example, $x \in \mathbb{R}$ means $x$ is a real-valued scalar.  
  - $x, y \in \{0, 1\}$ means $x$ and $y$ can only take values 0 or 1.  

#### Scalars as Tensors
- Scalars are implemented as tensors containing a single element.  
- We can perform basic operations like addition, multiplication, division, and exponentiation on scalars.
